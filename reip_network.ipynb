```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rc("figure", figsize=(15,8))
```

```python
df = pd.read_csv("sheet1.csv").set_index("Distance (ft)").sort_index()
df
```

```python
df.describe()
```

```python
#NanoStation Loco
```

```python
df["Upload (mbps)"].plot()
df["Down (mbps)"].plot()
plt.legend();
```

```python
plt.subplot(121)
df["Quality (%)"].plot(color="purple")
plt.legend()

plt.subplot(122)
df["Strength (dBm)"].plot(color="green")
plt.legend();
```
